# PLOTS

# Boxplots

```python
#comapring a numerical variable by categories

#comparing city mileage by car's transmission
```

```python
#using matlplotlib to adjust the figure size

plt.figure(figsize = (12,6))
sns.boxplot( x = 'Man.trans.avail' , y = 'MPGcity' , data = df)
```

# Stripplots

```python
#plotting with jitter

plt.figure(figsize = (12,6))
sns.stripplot(y = "MPGcity", x = "AirBags", data = df)
```

```python
#plotting without jitter

plt.figure(figsize = (12,6))
sns.stripplot(y = "MPGcity", x = "AirBags", data = df , jitter = False)
```

# Scatterplots

```python
#using matplotlib

plt.figure(figsize = (12,6))
plt.scatter(df.Horsepower , df.MPGcity , color = 'r' , s = 9)
plt.ylabel('MPGcity')
plt.xlabel('Horsepower');
```

```python
#scatter plot using seaborn

plt.figure(figsize = (12,6))
sns.regplot("Horsepower","MPGcity", data = df) 
```

```python
#lowess smoother

sns.regplot("Horsepower", "MPGcity", data = df , lowess = True) 
```

# Seaborn lmplot

```python
#scatterplot and regression lines for cars with different transmission
```

```python
sns.lmplot('Horsepower' , 'MPGcity' , data = df
           , hue = 'Man.trans.avail' , height = 7 , aspect = 1.5);
```

```python
#removing the confidence bounds

sns.lmplot('Horsepower' , 'MPGcity' , data = df
           , hue = 'Man.trans.avail' , height = 7 , aspect = 1.5 , ci = None);
```

```python
#using two facets using two categorical variables 

sns.lmplot('Horsepower' , 'MPGcity' , data = df
           , col = 'Man.trans.avail' , ci = None);
```

```python
#using four facets using four categorical variables 

sns.lmplot('Horsepower' , 'MPGcity' , data = df
           , col = 'Man.trans.avail' , row = 'Origin' , ci = None);
```

```python
#ploynomial regression with lmplot

sns.lmplot('Horsepower' , 'MPGcity' , data = df , height = 7
           , aspect = 1.5, ci = None);
```

```python
#ploynomial regression of order 2 with lmplot

sns.lmplot('Horsepower' , 'MPGcity' , data = df , height = 7
           , aspect = 1.5, ci = None , order = 2);
```
